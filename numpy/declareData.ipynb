# Declarate data functions

```python
import numpy as np
```

- **np.array** is a function used to create a NumPy array from an iterable object like a Python list.

```python
#Using python list
data_list = [1, 2, 3, 4, 5]
array_from_list = np.array(data_list)

print(f"Array created from list: {array_from_list} and the type is {type(array_from_list)}")
```

- **np.zeros** is a function used to create an array filled with zeros.

```python
#Using zeros function
array_zeros = np.zeros((2, 3))
print("\nArray created using zeros:")
print(array_zeros)
print(f"Type: {type(array_zeros)}")
```

- **np.ones** is a function used to create an array filled with ones.

```python
#Using ones function
array_ones = np.ones((3, 2))
print("\nArray created using ones:")
print(array_ones)
```

- **np.linspace** is a function used to create an array of evenly spaced numbers over a specified interval. 

```python
#Using linspace function
array_linspace = np.linspace(0, 10, 5)
print("\nArray created using linspace:")
print(array_linspace)
```

- **np.random.rand** is a function which generates random numbers from a uniform distribution over the interval [0, 1).

```python
#Using random.rand function
#Generate a 2D array of random numbers with shape (m, n)
array_random = np.random.rand(3, 3)
print("\nArray created using random.rand:")
print(array_random)
```

- **np.random.randint** is a function which generates random integers from a specified low (inclusive) to high (exclusive) range. 

```python
#Using random.randint function
# Generate a 2D array of random integers between 1 and 10 with shape (m, n)
array_randint = np.random.randint(1, 10, size=(2, 3))
print("\nArray created using random.randint:")
print(array_randint)
```

- **np.eye** is a function that creates a 2-dimensional array (or matrix) with diagonal elements set to 1 and off-diagonal elements set to 0.

```python
#Using eye function
array_eye = np.eye(3)
print("\nArray created using eye:")
print(array_eye)
# Generate a 3x5 identity matrix
identity_matrix_3x5 = np.eye(3, 5)
print("\nArray with 3x5 matrix:")
print(identity_matrix_3x5)
```

- **np.full** is a function that creates a new array filled with a specified value. It allows you to create an array of any shape and fill it with a constant value.

```python
#Using full function
array_full = np.full((2, 2), 7)
print("\nArray created using full:")
print(array_full)
```
